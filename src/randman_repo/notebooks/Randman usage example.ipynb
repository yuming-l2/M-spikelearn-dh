```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os, sys
```

```python
# Define 1D smooth random manifold in 2 space
rm = Randman(2,1)

# Get manifold points from intrinsic 1D manifold coordinates
x1 = np.linspace(0,1,100).reshape((-1,1))
y1 = rm.eval_manifold(x1)

# Randomly sample points from manifold
x2,y2 = rm.get_random_manifold_samples(100)

# Plot samples
p0=plt.scatter(y1[:,0],y1[:,1], label="Linspaced")
p0=plt.scatter(y2[:,0],y2[:,1], label="Randomly sampled")
plt.legend()
sns.despine()
```
