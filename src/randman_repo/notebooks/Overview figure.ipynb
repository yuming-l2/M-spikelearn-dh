```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
import seaborn as sns
import numpy as np
import os, sys
```

```python
sys.path.append("../")
import randman
from randman import Randman
```

```python
seed = 13
```

```python
# Illustrate effect of smoothness parameter \alpha

gs = GridSpec(1,4)
fig = plt.figure(figsize=(7,3),dpi=100)

for i,alpha in enumerate([3,2,1,0.5]):
    rm = Randman(3,1,alpha=alpha, seed=seed)
    x,y = rm.get_random_manifold_samples(1000)
    
    ax = fig.add_subplot(gs[i], projection='3d')
    ax.scatter(y[:,0], y[:,1], y[:,2], marker=".",alpha=0.1)
    ax.set_title(r"$\alpha=%g$"%alpha)
    
plt.tight_layout()
```

```python
# Vary number of random manifolds
alpha = 3.0

gs = GridSpec(2,4)
fig = plt.figure(figsize=(7,6),dpi=100)

for h in range(2):
    for i in range(4):
        ax = fig.add_subplot(gs[i+4*h], projection='3d')
        for j in range(i+1):
            D = h+1
            n = j+1
            rm = Randman(3,D,alpha=alpha,seed=seed+j)
            x,y = rm.get_random_manifold_samples(1000)
            ax.scatter(y[:,0], y[:,1], y[:,2], marker=".",alpha=0.1)
        ax.set_title(r"$n=%i, D=%i, \alpha=%g$"%(n,D,alpha))
    
plt.tight_layout()
```
